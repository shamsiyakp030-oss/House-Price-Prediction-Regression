### Objective

The objective of this project is to apply regression algorithms to predict house prices based on features like bedrooms, bathrooms, area, etc.

### Dataset
* Dataset: House Sales Prediction
* Source: Kaggle
Link: https://www.kaggle.com/datasets/harlfoxem/housesalesprediction
### Data Loading & Preprocessing (2 Marks)
#### Steps Performed:
* Loaded dataset using pandas
* Checked missing values using .isnull().sum()
* Removed or handled missing values
* Dropped unnecessary columns (like ID, date if not useful)
* Scaled features using StandardScaler
* Handled outliers using boxplot & IQR method
### Goal
To predict house prices using regression models based on features like area, bedrooms, and bathrooms.

## code

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("kc_house_data.csv")

# Check missing values
print(df.isnull().sum())

# Drop unnecessary columns
df = df.drop(['id', 'date'], axis=1)

# Features and target
X = df.drop('price', axis=1)
y = df['price']

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

* Cleaning ensures data is accurate
* Scaling helps models perform better
* Removing outliers improves prediction accurac

## Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
df['price'].hist()
plt.title("Price Distribution")
plt.show()

# Scatter plot
plt.scatter(df['sqft_living'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("price Distribution")
plt.show()

# Boxplot
sns.boxplot(x=df['price'])
plt.title("price Distribution")
plt.show()

# Correlation
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("price Distribution")
plt.show()
```

#### Insights:
* Price is right-skewed
* Area has strong relationship with price
* Some outliers exist in price

## Model Building

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

### 1.Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

* House prices often depend linearly on features like area

### 2. Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
```

* Handles complex relationships in housing data

### 3. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

* More accurate and stable for large datasets

### 4. Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate(y_test, y_pred):
    print("MAE:", mean_absolute_error(y_test, y_pred))
    print("MSE:", mean_squared_error(y_test, y_pred))
    print("R2 Score:", r2_score(y_test, y_pred))
    print()

print("Linear Regression:")
evaluate(y_test, y_pred_lr)

print("Decision Tree:")
evaluate(y_test, y_pred_dt)

print("Random Forest:")
evaluate(y_test, y_pred_rf)
```

## 6. Conclusion:
* Regression models can effectively predict house prices
* Random Forest gave the most accurate results
* Proper preprocessing improves model performance
